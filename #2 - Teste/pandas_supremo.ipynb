```python
import pandas as pd
import numpy as np

df_vendas = pd.read_csv('vendas_tech.csv', low_memory=False)
display(df_vendas)
df_gerentes = pd.read_excel('gerentes_lojas.xlsx')
display(df_gerentes)
```

```python
# inspecao

display(df_vendas)
display(df_vendas.head(15))
display(df_vendas.columns)
display(df_vendas.info())
```

```python
# tratamento de dados
# display(df_vendas['Loja'])
df_novo = df_vendas[['Loja', 'Produto']]
# display(df_novo)

# colunas
df_analise = df_vendas.drop(columns=['Data_Base'])

# nulos
# df_analise = df_analise.dropna() exclui todas linhas com pelo menos 1 valor vazio
df_analise["Loja"] = df_analise["Loja"].fillna("Sem Loja") # preenche os valores vazios com "Sem Loja"

# tipo de dados
df_analise['Data'] = pd.to_datetime(df_analise['Data'], format="%Y-%m-%d")

# padronizacao
df_analise['Loja'] = df_analise['Loja'].str.strip()
df_analise['Loja'] = df_analise['Loja'].str.title()

# duplicatas
df_analise = df_analise.drop_duplicates(subset=['ID_Pedido', 'Loja'], keep='first')

display(df_analise)
display(df_analise.info())
```
