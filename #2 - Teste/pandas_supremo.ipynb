```python
import pandas as pd
import numpy as np

df_vendas = pd.read_csv('vendas_tech.csv', low_memory=False)
display(df_vendas)
df_gerentes = pd.read_excel('gerentes_lojas.xlsx')
display(df_gerentes)
```

```python
# inspecao

display(df_vendas)
display(df_vendas.head(15))
display(df_vendas.columns)
display(df_vendas.info())
```

```python
# tratamento de dados
# display(df_vendas['Loja'])
df_novo = df_vendas[['Loja', 'Produto']]
# display(df_novo)

# colunas
df_analise = df_vendas.drop(columns=['Data_Base'])

# nulos
# df_analise = df_analise.dropna() exclui todas linhas com pelo menos 1 valor vazio
df_analise["Loja"] = df_analise["Loja"].fillna("Sem Loja") # preenche os valores vazios com "Sem Loja"

# tipo de dados
df_analise['Data'] = pd.to_datetime(df_analise['Data'], format="%Y-%m-%d")

# padronizacao
df_analise['Loja'] = df_analise['Loja'].str.strip()
df_analise['Loja'] = df_analise['Loja'].str.title()

# encadeamento
# df_analise["Loja"] = df_analise["Loja"].fillna("Sem Loja").str.strip().str.title()

# duplicatas
df_analise = df_analise.drop_duplicates(subset=['ID_Pedido', 'Loja'], keep='first').reset_index(drop=True)

display(df_analise)
display(df_analise.info())
```

```python
# criar novas colunas

# faturamento
df_analise['Faturamento'] = df_analise['Qtd'] * df_analise['Preco_Unitario']

# forma de venda
df_analise['Forma_de_venda'] = np.where(df_analise['Loja'] == 'Sem Loja', 'Online', 'Presencial')

# regiao
display(df_analise['Loja'].unique())

dic_regioes = {
    'São Paulo': 'Sudeste',
    'Belo Horizonte': 'Sudeste',
    'Sem Loja': 'Online',
    'Rio De Janeiro': 'Sudeste',
    'Salvador': 'Nordeste',
    'Recife': 'Nordeste',
    'Curitiba': 'Sul',
    'Porto Alegre': 'Sul'
}
df_analise['Regiao'] = df_analise['Loja'].map(dic_regioes)

display(df_analise)
```

```python
# analisar - filtrar
df_analise = df_analise.sort_values(by=['Data', 'Faturamento'], ascending=True)
df_analise = df_analise.reset_index(drop=True) # drop joga o indice antigo fora

# .loc: id do pedido -> localiza por nome da coluna e nome da linha
id_pedido = 4

loja = df_analise.loc[3, 'Loja'] # nome do indice, coluna
cliente = df_analise.loc[3, 'Cliente']
produto = df_analise.loc[3, 'Produto']
# print(loja, cliente, produto)

# .iloc -> por posicao
id_pedido = 4
loja = df_analise.iloc[3, 2] # posicao, numero da coluna
produto = df_analise.iloc[3, 3]
cliente = df_analise.iloc[3, 6]
# print(loja, cliente, produto)

# condicional
df_id_pedido_4 = df_analise[df_analise['ID_Pedido'] == 4]

# exportar pedacos da base
df_vendas_sp = df_analise[df_analise['Loja'] == 'São Paulo']
df_vendas_sp.to_csv('Vendas_SP', index=False)
# display(df_vendas_sp)

# exportas vendas de 2024
df_vendas_2024 = df_analise[df_analise['Data'] >= '2024-01-01']

# duplas condicoes
condicao = df_analise['Produto'] == 'Cabo HDMI'
condicao2 = df_analise['Regiao'] == 'Sul'

df_vendas_hdmi_sul = df_analise[(condicao) & (condicao2)] # para usar ou é |
display(df_vendas_hdmi_sul)

display(df_vendas_2024)
display(df_analise)
```

```python
# analiser de agrupamento
analise_loja = df_analise[['Loja', 'Faturamento']].groupby('Loja').sum()
analise_loja = analise_loja.sort_values(by='Faturamento', ascending=False)
analise_loja = analise_loja.reset_index()
analise_loja['Faturamento'] = analise_loja['Faturamento'].map('R$ {:,.2f}'.format)

display(analise_loja)

# ranking de produtos online
df_vendas_online = df_analise[df_analise['Loja'] == 'Sem Loja']
analise_produtos_online = df_vendas_online[['Produto', 'Qtd']].groupby('Produto').sum().sort_values(by='Qtd', ascending=False)

analise_produtos_online = analise_produtos_online.rename(columns={'Qtd': 'Vendas totais'})
display(analise_produtos_online)

# análise de ranking por loja e produto
# quais produtos venderam mais em quais lojas
analise_produtos_em_lojas = df_analise[['Loja', 'Produto', 'Qtd']].groupby(['Loja', 'Produto']).sum()

with pd.option_context('display.max_rows', None):
    display(analise_produtos_em_lojas)
```
